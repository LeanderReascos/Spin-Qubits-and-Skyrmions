```python
import sympy as sp
```

```python
Bsk = sp.symbols('B_\\mathrm{sk}')  # stray field from skyrmion

D, L = sp.symbols('D L', real=True, positive=True)  # diameter and length of the cylinder
Dx, Dy, Dz = sp.symbols('Delta_x Delta_y Delta_z', real=True)  # distance from the cylinder center to the point of interest

R = sp.Rational(1, 2) * D  # radius of the cylinder
r = sp.symbols('r', real=True, positive=True)  # radial distance from the cylinder axis to the point of interest
z = -Dz

theta = sp.symbols('theta', real=True)  # angle in the cylinder cross section

xi_p = sp.symbols('xi_+', real=True)  # auxiliary variable    2 * Dz + L
xi_m = sp.symbols('xi_-', real=True)  # auxiliary variable    2 * Dz - L

```

```python
Br_p = - 2 * Bsk * R * sp.integrate(sp.cos(theta) / sp.sqrt((-sp.Rational(1,2) * xi_m)**2 + R**2 - 2*R*r*sp.cos(theta) + r**2).factor(), (theta, 0, sp.pi))
Br_m = - 2 * Bsk * R * sp.integrate(sp.cos(theta) / sp.sqrt((-sp.Rational(1,2) * xi_p)**2 + R**2 - 2*R*r*sp.cos(theta) + r**2).factor(), (theta, 0, sp.pi))

Br_p
```
