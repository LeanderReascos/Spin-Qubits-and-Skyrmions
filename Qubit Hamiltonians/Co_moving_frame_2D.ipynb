```python
dim_spin = 2

Spin = RDBasis("\\sigma", 'spin', dim = dim_spin)
s0, sx, sy, sz = Spin._basis

a_1 = RDBoson("a_1", subspace ="boson1")
ad_1 = RDBoson("{a_1^\\dagger}", subspace ="boson1", is_annihilation=False)
a_2 = RDBoson("a_2", subspace ="boson2")
ad_2 = RDBoson("{a_2^\\dagger}", subspace ="boson2", is_annihilation=False)

commutation_relations = {
    a_1 * ad_1 : ad_1 * a_1 + 1,
    a_2 * ad_2 : ad_2 * a_2 + 1,
    a_2 * ad_1 : ad_1 * a_2,
    ad_2 * a_1 : a_1 * ad_2,
    ad_2 * ad_1 : ad_1 * ad_2,
    a_2 * a_1 : a_1 * a_2
}

hbar = RDsymbol("hbar", order = 0)
omega_1 = RDsymbol("omega_1", order = 0)
omega_2 = RDsymbol("omega_2", order = 0)
Omega_z = RDsymbol("\\Omega_{z}", order = 0)

H0 = hbar * omega_1 * ad_1 * a_1 + hbar * omega_2 * ad_2 * a_2 - sp.Rational(1,2) * hbar * Omega_z * sz
display(H0)

f1 = RDsymbol("f_1", order = 1)
f2 = RDsymbol("f_2", order = 1)
g1 = RDsymbol("g_1", order = 1)
g2 = RDsymbol("g_2", order = 1)

V1 = hbar * f1 * (ad_1 + a_1) * sx - I * hbar * g2 * (ad_2 - a_2) * sx + hbar * f2 * (ad_2 + a_2) * sy - I * hbar * g1 * (ad_1 - a_1) * sy
display(V1)

delta_1 = RDsymbol("\\delta_1", order = 1)
delta_2 = RDsymbol("\\delta_2", order = 1)

delta_m_1 = 0#RDsymbol("\\delta_{-1}", order = 2)
delta_m_2 = 0#RDsymbol("\\delta_{-2}", order = 2)

H2 = sp.Rational(1, 2) * hbar * delta_1 *sz + sp.Rational(1, 2) * hbar * delta_2 *sz + hbar * delta_1 * ad_1 * a_1 * sz + hbar * delta_2 * ad_2 * a_2 * sz
display(H2)

tp = RDsymbol("t_{+}", order = 1)
tm = 0#RDsymbol("t_{-}", order = 2)

V2 = sp.Rational(1,2) * hbar * delta_m_1 * (ad_1**2 + a_1**2) * sz + sp.Rational(1,2) * hbar * delta_m_2 * (ad_2**2 + a_2**2) * sz + I * hbar * tp * (ad_1*ad_2 - a_1*a_2) * sz + I * hbar * tm * (ad_1*a_2 - a_1*ad_2) * sz
display(V2)

H = H0 + V1 + H2 + V2
```

```python
sol = solver(H, Spin, order=2, full_diagonal=True, commutation_relations=commutation_relations)
sol[0]
```

```python
result_dict_aux = group_by_infinite_operators(sol[0] - H0 - H2)
result_dict = group_by_finite_operators(result_dict_aux[1])
result_dict.pop(1)
result_dict_aux.pop(1)

subs_dict = {
    ((omega_1**2 - Omega_z**2) * (omega_2**2 - Omega_z**2)).expand() :(omega_1**2 - Omega_z**2) * (omega_2**2 - Omega_z**2)
}

refine_1 = sp.Q.positive(omega_1**2 - Omega_z**2)
refine_2 = sp.Q.positive(omega_2**2 - Omega_z**2)
refine = refine_1 & refine_2

num_expr, den_expr =  result_dict[sz].subs(subs_dict).together().refine(refine_1).refine(refine_2).as_numer_denom()
num_expr = num_expr.simplify().expand().collect([Omega_z*f1**2, Omega_z*g1**2, 2*omega_1*f1*g1, 2*omega_2*f2*g2, Omega_z*f2**2, Omega_z*g2**2]).collect([hbar*omega_1**2 - hbar*Omega_z**2, hbar*omega_2**2 - hbar*Omega_z**2], evaluate=False)
den_expr = den_expr.collect([omega_1**2 - Omega_z**2, omega_2**2 - Omega_z**2], evaluate=False)


result_dict[sz] = (hbar * list(num_expr.values())[0]/list(den_expr.keys())[0] + hbar * list(num_expr.values())[1]/list(den_expr.values())[0]).refine(refine)

H2_final = H0 + H2 + result_dict[sz] * sz

for key, val in result_dict_aux.items():
    for key2, val2 in group_by_finite_operators(val).items():
        if val2 != 0:
            result_dict[key * key2] = val2.simplify().expand().together()
            H2_final += key * key2 *  val2.simplify().expand().together()
        
display_dict(result_dict)
```

## Add the driving term

```python
d = RDsymbol("d", order = 0)
D = RDsymbol("D", order = 0)

mu1 = RDsymbol("\\mu_1", order = 1)
mu2 = RDsymbol("\\mu_2", order = 1)
eta1 = RDsymbol("\\eta_1", order = 1)
eta2 = RDsymbol("\\eta_2", order = 1)

Hd = -d * (mu1 * (ad_1 + a_1) + I * eta2 * (ad_2 - a_2))
HD = -D * (mu2 * (ad_2 + a_2) + I * eta1 * (ad_1 - a_1))

H_driving = Hd + HD
display(H_driving)
```

```python
driving_term_dict = group_by_list_operators(sp.Add(*[k*v for k,v in expand_commutator(Commutator(H_driving, S1)).doit().refine(refine).collect([(omega_1**2 - Omega_z**2), (omega_2**2 - Omega_z**2)], evaluate=False).items()]).collect([D,d]), [sx, sy, D, d])
display_dict(driving_term_dict)

driving_term = sp.Add(*[k*v for k,v in expand_commutator(Commutator(H_driving, S1)).doit().refine(refine).collect([(omega_1**2 - Omega_z**2), (omega_2**2 - Omega_z**2)], evaluate=False).items()]).collect([D,d])
driving_term
```

```python
def subs_in_dict(dict_to_subs, subs):
    dict_sub = {}
    for key, val in dict_to_subs.items():
        dict_sub[key] = val.subs(subs).expand()
    return dict_sub
```
